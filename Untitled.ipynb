```python
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
stocks = pd.read_csv('SP500_Yahoo_GSPC.csv')
print(stocks.head(5))
```

```python
date = stocks['Date'].values
close = stocks['Close'].values

startIndex = np.where(date == '1980-01-02')[0]
if len(startIndex) == 0:
    print('err: date invalid (market not open?)')
startIndex = startIndex[0]
    
date = date[startIndex:]
close = close[startIndex:]
print(date)
print(close)
```

```python
slack = 0.02
isMax = []
maxVal = -1
for i in range(len(close)):
    if maxVal * (1 - slack) <= close[i]:
        isMax.append(i)
    if close[i] > maxVal:
        maxVal = close[i]

print(isMax[:10])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.plot(range(len(date)), close, 'b-')
for i in isMax:
    plt.axvspan(i, i + 1, color='red', alpha=0.5)
plt.show()
print(1 - len(isMax) / len(date))
```
